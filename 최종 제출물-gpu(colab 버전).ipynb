```python
# ë‚œì ì¶œì²˜ ì•Œ ìˆ˜ ì—†ìŒ -> ë³¸ì¸ì œê³µ replace
train_df['ë‚œì ì¶œì²˜'] = train_df['ë‚œì ì¶œì²˜'].replace('ì•Œ ìˆ˜ ì—†ìŒ', 'ë³¸ì¸ ì œê³µ')
test_df['ë‚œì ì¶œì²˜'] = test_df['ë‚œì ì¶œì²˜'].replace('ì•Œ ìˆ˜ ì—†ìŒ', 'ë³¸ì¸ ì œê³µ')
```

```python
# Unknown
train_df['íŠ¹ì • ì‹œìˆ  ìœ í˜•'] = train_df['íŠ¹ì • ì‹œìˆ  ìœ í˜•'].fillna('Unknown')
test_df['íŠ¹ì • ì‹œìˆ  ìœ í˜•'] = test_df['íŠ¹ì • ì‹œìˆ  ìœ í˜•'].fillna('Unknown')

# íŠ¹ì • ì»¬ëŸ¼ ë¨¼ì € ëŒ€ì²´
s_col = ['PGD ì‹œìˆ  ì—¬ë¶€', 'ì°©ìƒ ì „ ìœ ì „ ê²€ì‚¬ ì‚¬ìš© ì—¬ë¶€', 'PGS ì‹œìˆ  ì—¬ë¶€']
train_df[s_col] = train_df[s_col].fillna(0)
test_df[s_col] = test_df[s_col].fillna(0)

# IVFê°€ ì•„ë‹Œ DIì˜ ê²½ìš° ê²°ì¸¡ì¹˜ ëŒ€ì²´
# ì„¤ë¬¸ ì¡°ì‚¬ì—ì„œ DIì˜ ê²½ìš° ëŒ€ë‹µí•  í•„ìš”ê°€ ì—†ëŠ” í•­ëª©ë“¤ì´ ê²°ì¸¡ì¹˜ë¡œ ë“¤ì–´ê°„ ë“¯
columns_to_update = ['ë‹¨ì¼ ë°°ì•„ ì´ì‹ ì—¬ë¶€', 'ì°©ìƒ ì „ ìœ ì „ ì§„ë‹¨ ì‚¬ìš© ì—¬ë¶€', 'ë°°ì•„ ìƒì„± ì£¼ìš” ì´ìœ ',
                     'ì´ ìƒì„± ë°°ì•„ ìˆ˜', 'ë¯¸ì„¸ì£¼ì…ëœ ë‚œì ìˆ˜', 'ë¯¸ì„¸ì£¼ì…ì—ì„œ ìƒì„±ëœ ë°°ì•„ ìˆ˜',
                     'ì´ì‹ëœ ë°°ì•„ ìˆ˜', 'ë¯¸ì„¸ì£¼ì… ë°°ì•„ ì´ì‹ ìˆ˜', 'ì €ì¥ëœ ë°°ì•„ ìˆ˜','ë¯¸ì„¸ì£¼ì… í›„ ì €ì¥ëœ ë°°ì•„ ìˆ˜',
                     'í•´ë™ëœ ë°°ì•„ ìˆ˜', 'í•´ë™ ë‚œì ìˆ˜', 'ìˆ˜ì§‘ëœ ì‹ ì„  ë‚œì ìˆ˜', 'ì €ì¥ëœ ì‹ ì„  ë‚œì ìˆ˜',
                     'í˜¼í•©ëœ ë‚œì ìˆ˜', 'íŒŒíŠ¸ë„ˆ ì •ìì™€ í˜¼í•©ëœ ë‚œì ìˆ˜', 'ê¸°ì¦ì ì •ìì™€ í˜¼í•©ëœ ë‚œì ìˆ˜',
                     'ë™ê²° ë°°ì•„ ì‚¬ìš© ì—¬ë¶€', 'ì‹ ì„  ë°°ì•„ ì‚¬ìš© ì—¬ë¶€', 'ê¸°ì¦ ë°°ì•„ ì‚¬ìš© ì—¬ë¶€', 'ëŒ€ë¦¬ëª¨ ì—¬ë¶€']

# train_df ì²˜ë¦¬
for column in columns_to_update:
    if train_df[column].dtype == 'object':
        # object íƒ€ì…ì´ë©´ 'Not Answer(DI)'ë¡œ ì±„ìš°ê¸°
        train_df[column] = train_df[column].fillna('Not Answer(DI)')
    elif train_df[column].dtype in ['float64', 'int64']:
        # ìˆ«ì íƒ€ì…ì´ë©´ 0ë¡œ ì±„ìš°ê¸°
        train_df[column] = train_df[column].fillna(0)

# test_df ì²˜ë¦¬
for column in columns_to_update:
    if test_df[column].dtype == 'object':
        # object íƒ€ì…ì´ë©´ 'Not Answer(DI)'ë¡œ ì±„ìš°ê¸°
        test_df[column] = test_df[column].fillna('Not Answer(DI)')
    elif test_df[column].dtype in ['float64', 'int64']:
        # ìˆ«ì íƒ€ì…ì´ë©´ 0ë¡œ ì±„ìš°ê¸°
        test_df[column] = test_df[column].fillna(0)

# ê²½ê³¼ì¼ ì»¬ëŸ¼ ëŒ€ì²´
o_col = ['ë‚œì ì±„ì·¨ ê²½ê³¼ì¼']
train_df[o_col] = train_df[o_col].fillna(1)
test_df[o_col] = test_df[o_col].fillna(1)

d_col = ['ë‚œì í˜¼í•© ê²½ê³¼ì¼', 'ë°°ì•„ ì´ì‹ ê²½ê³¼ì¼', 'ë°°ì•„ í•´ë™ ê²½ê³¼ì¼']
train_df[d_col] = train_df[d_col].fillna(999)
test_df[d_col] = test_df[d_col].fillna(999)
```

```python
# íšŸìˆ˜, íšŒ íƒ€ì… ë³€ê²½
int_col = ['ì´ ì‹œìˆ  íšŸìˆ˜','í´ë¦¬ë‹‰ ë‚´ ì´ ì‹œìˆ  íšŸìˆ˜', 'IVF ì‹œìˆ  íšŸìˆ˜', 'DI ì‹œìˆ  íšŸìˆ˜',
           'ì´ ì„ì‹  íšŸìˆ˜', 'IVF ì„ì‹  íšŸìˆ˜','DI ì„ì‹  íšŸìˆ˜', 'ì´ ì¶œì‚° íšŸìˆ˜', 'IVF ì¶œì‚° íšŸìˆ˜',
           'DI ì¶œì‚° íšŸìˆ˜']

for col in int_col:
    train_df[col] = train_df[col].astype(str).str.extract('(\d+)').astype(int)
    test_df[col] = test_df[col].astype(str).str.extract('(\d+)').astype(int)
```

```python
# ê°’ì´ 1ì¸ í•­ëª©ì˜ ì´ë¦„ ì¶”ì¶œ
to_drop = train_df['ë°°ë€ ìœ ë„ ìœ í˜•'].value_counts()[train_df['ë°°ë€ ìœ ë„ ìœ í˜•'].value_counts() == 1].index

# í•´ë‹¹ í•­ëª© ì œê±° (train_dfë§Œ)
train_df = train_df[~train_df['ë°°ë€ ìœ ë„ ìœ í˜•'].isin(to_drop)]
```

```python
# ì¡°ê±´ : ë‚œì ì¶œì²˜ == "ë³¸ì¸ ì œê³µ" â†’ ë‚œì ê¸°ì¦ì ë‚˜ì´ë¥¼ ì‹œìˆ  ë‹¹ì‹œ ë‚˜ì´ë¡œ ì„¤ì •
condition2 = (train_df['ë‚œì ì¶œì²˜'] == "ë³¸ì¸ ì œê³µ")
train_df.loc[condition2, 'ë‚œì ê¸°ì¦ì ë‚˜ì´'] = train_df.loc[condition2, 'ì‹œìˆ  ë‹¹ì‹œ ë‚˜ì´']

condition2_t = (test_df['ë‚œì ì¶œì²˜'] == "ë³¸ì¸ ì œê³µ")
test_df.loc[condition2_t, 'ë‚œì ê¸°ì¦ì ë‚˜ì´'] = test_df.loc[condition2_t, 'ì‹œìˆ  ë‹¹ì‹œ ë‚˜ì´']
```

```python
# ì»¬ëŸ¼ drop
drop_col = ['ì„ì‹  ì‹œë„ ë˜ëŠ” ë§ˆì§€ë§‰ ì„ì‹  ê²½ê³¼ ì—°ìˆ˜', 'ë‚œì í•´ë™ ê²½ê³¼ì¼', 'ë¶ˆì„ ì›ì¸ - ì •ì í˜•íƒœ', 'ë¶ˆì„ ì›ì¸ - ì •ì ìš´ë™ì„±',
            'ë¶ˆì„ ì›ì¸ - ì •ì ë©´ì—­í•™ì  ìš”ì¸', 'ë¶ˆì„ ì›ì¸ - ì •ì ë†ë„', 'ë¶ˆì„ ì›ì¸ - ìê¶ê²½ë¶€ ë¬¸ì œ', 'ë¶ˆì„ ì›ì¸ - ì—¬ì„± ìš”ì¸',
            'ëŒ€ë¦¬ëª¨ ì—¬ë¶€', 'ë¶€ë¶€ ë¶€ ë¶ˆì„ ì›ì¸']


train_df.drop(drop_col, axis=1, inplace=True)
test_df.drop(drop_col, axis=1, inplace=True)
```

## íŒŒìƒë³€ìˆ˜

```python
# ì‹œìˆ  ë°›ì•˜ëŠ”ë° ì„ì‹ ì„ ëª»í•œ ê²½ìš° = ë‚œì„
train_df['ë‚œì„ ì—¬ë¶€'] = train_df['ì´ ì‹œìˆ  íšŸìˆ˜'] - train_df['ì´ ì„ì‹  íšŸìˆ˜']
test_df['ë‚œì„ ì—¬ë¶€'] = test_df['ì´ ì‹œìˆ  íšŸìˆ˜'] - test_df['ì´ ì„ì‹  íšŸìˆ˜']

# ì„ì‹ ì„ í–ˆëŠ”ë° ì¶œì‚°ì„ ëª»í•œê²½ìš° = ìœ ì‚°
train_df['ìœ ì‚° ì—¬ë¶€'] = train_df['ì´ ì„ì‹  íšŸìˆ˜'] - train_df['ì´ ì¶œì‚° íšŸìˆ˜']
train_df['ìœ ì‚° ì—¬ë¶€'] = train_df['ìœ ì‚° ì—¬ë¶€'].apply(lambda x: 1 if x > 0 else 0)

test_df['ìœ ì‚° ì—¬ë¶€'] = test_df['ì´ ì„ì‹  íšŸìˆ˜'] - test_df['ì´ ì¶œì‚° íšŸìˆ˜']
test_df['ìœ ì‚° ì—¬ë¶€'] = test_df['ìœ ì‚° ì—¬ë¶€'].apply(lambda x: 1 if x > 0 else 0)
```

```python
# ë¯¸ì„¸ì£¼ì…(ICSI) ì•„ë‹Œ IVF
train_df['ë¯¸ì„¸ì£¼ì…ì´ ì•„ë‹Œ ë°°ì•„ ì´ì‹ ìˆ˜'] = train_df['ì´ì‹ëœ ë°°ì•„ ìˆ˜'] - train_df['ë¯¸ì„¸ì£¼ì… ë°°ì•„ ì´ì‹ ìˆ˜']
test_df['ë¯¸ì„¸ì£¼ì…ì´ ì•„ë‹Œ ë°°ì•„ ì´ì‹ ìˆ˜'] = test_df['ì´ì‹ëœ ë°°ì•„ ìˆ˜'] - test_df['ë¯¸ì„¸ì£¼ì… ë°°ì•„ ì´ì‹ ìˆ˜']

train_df['ë¯¸ì„¸ì£¼ì…ì´ ì•„ë‹Œ ë°°ì•„ ìƒì„± ìˆ˜'] = train_df['ì´ ìƒì„± ë°°ì•„ ìˆ˜'] - train_df['ë¯¸ì„¸ì£¼ì…ì—ì„œ ìƒì„±ëœ ë°°ì•„ ìˆ˜']
test_df['ë¯¸ì„¸ì£¼ì…ì´ ì•„ë‹Œ ë°°ì•„ ìƒì„± ìˆ˜'] = test_df['ì´ ìƒì„± ë°°ì•„ ìˆ˜'] - test_df['ë¯¸ì„¸ì£¼ì…ì—ì„œ ìƒì„±ëœ ë°°ì•„ ìˆ˜']

train_df['ë¯¸ì„¸ì£¼ì…ì´ ì•„ë‹Œ ë°°ì•„ ì €ì¥ ìˆ˜'] = train_df['ì €ì¥ëœ ë°°ì•„ ìˆ˜'] - train_df['ë¯¸ì„¸ì£¼ì… í›„ ì €ì¥ëœ ë°°ì•„ ìˆ˜']
test_df['ë¯¸ì„¸ì£¼ì…ì´ ì•„ë‹Œ ë°°ì•„ ì €ì¥ ìˆ˜'] = test_df['ì €ì¥ëœ ë°°ì•„ ìˆ˜'] - test_df['ë¯¸ì„¸ì£¼ì… í›„ ì €ì¥ëœ ë°°ì•„ ìˆ˜']
```

```python
train_df['í´ë¦¬ë‹‰ ì™¸ ì´ ì‹œìˆ  íšŸìˆ˜'] = train_df['ì´ ì‹œìˆ  íšŸìˆ˜'] - train_df['í´ë¦¬ë‹‰ ë‚´ ì´ ì‹œìˆ  íšŸìˆ˜']
test_df['í´ë¦¬ë‹‰ ì™¸ ì´ ì‹œìˆ  íšŸìˆ˜'] = test_df['ì´ ì‹œìˆ  íšŸìˆ˜'] - test_df['í´ë¦¬ë‹‰ ë‚´ ì´ ì‹œìˆ  íšŸìˆ˜']

train_df['PGS ê²€ì‚¬ë¥¼ ë°›ê³ ë„ ì•ˆí•œ ì‚¬ëŒ'] = train_df['ì°©ìƒ ì „ ìœ ì „ ê²€ì‚¬ ì‚¬ìš© ì—¬ë¶€'] - train_df['PGS ì‹œìˆ  ì—¬ë¶€']
test_df['PGS ê²€ì‚¬ë¥¼ ë°›ê³ ë„ ì•ˆí•œ ì‚¬ëŒ'] = test_df['ì°©ìƒ ì „ ìœ ì „ ê²€ì‚¬ ì‚¬ìš© ì—¬ë¶€'] - test_df['PGS ì‹œìˆ  ì—¬ë¶€']
```

## ë°°ì•„ ë³€ìˆ˜

```python
# ë‹¨ì¼ ë°°ì•„ë¥¼ ì´ì‹í•œ ê²½ìš°ì˜ ì´ì‹ëœ ë°°ì•„ ìˆ˜ê°€ 1ì´ë©´ ì„±ê³µë¥ ì´ ë” ë†’ìŒ
train_df.loc[(train_df['ë‹¨ì¼ ë°°ì•„ ì´ì‹ ì—¬ë¶€'] == 1) & (train_df['ì´ì‹ëœ ë°°ì•„ ìˆ˜'] == 1), 'ì´ì‹ëœ ë°°ì•„ ìˆ˜'] = 1.5
test_df.loc[(test_df['ë‹¨ì¼ ë°°ì•„ ì´ì‹ ì—¬ë¶€'] == 1) & (test_df['ì´ì‹ëœ ë°°ì•„ ìˆ˜'] == 1), 'ì´ì‹ëœ ë°°ì•„ ìˆ˜'] = 1.5

train_df.loc[(train_df['ì°©ìƒ ì „ ìœ ì „ ê²€ì‚¬ ì‚¬ìš© ì—¬ë¶€'] == 1) & (train_df['PGS ì‹œìˆ  ì—¬ë¶€'] == 0), 'PGS ì‹œìˆ  ì—¬ë¶€'] = -1
test_df.loc[(test_df['ì°©ìƒ ì „ ìœ ì „ ê²€ì‚¬ ì‚¬ìš© ì—¬ë¶€'] == 1) & (test_df['PGS ì‹œìˆ  ì—¬ë¶€'] == 0), 'PGS ì‹œìˆ  ì—¬ë¶€'] = -1
```

```python
# ì´ì‹ ë°°ì•„ ìˆ˜/ ë‚˜ì´
# ë‚˜ì´ ì¬ ì •ì˜
age_bin_mapping = {
    "ë§Œ18-34ì„¸": 1,
    "ë§Œ35-37ì„¸": 2,
    "ë§Œ38-39ì„¸": 3,
    "ë§Œ40-42ì„¸": 4,
    "ë§Œ43-44ì„¸": 5,
    "ë§Œ45-50ì„¸": 6,
    "ì•Œ ìˆ˜ ì—†ìŒ": 7
}

# ë‚˜ì´  ì ìš©
train_df['ë‚˜ì´'] = train_df['ì‹œìˆ  ë‹¹ì‹œ ë‚˜ì´'].map(age_bin_mapping)
test_df['ë‚˜ì´'] = test_df['ì‹œìˆ  ë‹¹ì‹œ ë‚˜ì´'].map(age_bin_mapping)

# ë‚˜ì´ ë³„ ë°°ì•„ìˆ˜
train_df['ì´ì‹ ë°°ì•„ ìˆ˜/ë‚˜ì´'] = train_df['ì´ì‹ëœ ë°°ì•„ ìˆ˜'] / train_df['ë‚˜ì´']

test_df['ì´ì‹ ë°°ì•„ ìˆ˜/ë‚˜ì´'] = test_df['ì´ì‹ëœ ë°°ì•„ ìˆ˜'] / test_df['ë‚˜ì´']

# ë‚˜ì´ drop
train_df.drop('ë‚˜ì´', axis=1, inplace=True)
test_df.drop('ë‚˜ì´', axis=1, inplace=True)
```

## ë²”ì£¼í˜•

```python
# categoricalë¡œ ë°”ê¿”ì¤Œ
cat_col = ['ë°°ë€ ìê·¹ ì—¬ë¶€', 'ë‹¨ì¼ ë°°ì•„ ì´ì‹ ì—¬ë¶€', 'ì°©ìƒ ì „ ìœ ì „ ê²€ì‚¬ ì‚¬ìš© ì—¬ë¶€', 'ì°©ìƒ ì „ ìœ ì „ ì§„ë‹¨ ì‚¬ìš© ì—¬ë¶€',
 'ë‚¨ì„± ì£¼ ë¶ˆì„ ì›ì¸', 'ë‚¨ì„± ë¶€ ë¶ˆì„ ì›ì¸', 'ì—¬ì„± ì£¼ ë¶ˆì„ ì›ì¸', 'ì—¬ì„± ë¶€ ë¶ˆì„ ì›ì¸', 'ë¶€ë¶€ ì£¼ ë¶ˆì„ ì›ì¸',
  'ë¶ˆëª…í™• ë¶ˆì„ ì›ì¸', 'ë¶ˆì„ ì›ì¸ - ë‚œê´€ ì§ˆí™˜', 'ë¶ˆì„ ì›ì¸ - ë‚¨ì„± ìš”ì¸', 'ë¶ˆì„ ì›ì¸ - ë°°ë€ ì¥ì• ',
 'ë¶ˆì„ ì›ì¸ - ìê¶ë‚´ë§‰ì¦', 'ì´ ì‹œìˆ  íšŸìˆ˜', 'í´ë¦¬ë‹‰ ë‚´ ì´ ì‹œìˆ  íšŸìˆ˜', 'IVF ì‹œìˆ  íšŸìˆ˜', 'DI ì‹œìˆ  íšŸìˆ˜', 'ì´ ì„ì‹  íšŸìˆ˜',
 'IVF ì„ì‹  íšŸìˆ˜', 'DI ì„ì‹  íšŸìˆ˜', 'ì´ ì¶œì‚° íšŸìˆ˜', 'IVF ì¶œì‚° íšŸìˆ˜', 'DI ì¶œì‚° íšŸìˆ˜', 'ë™ê²° ë°°ì•„ ì‚¬ìš© ì—¬ë¶€',
 'ì‹ ì„  ë°°ì•„ ì‚¬ìš© ì—¬ë¶€', 'ê¸°ì¦ ë°°ì•„ ì‚¬ìš© ì—¬ë¶€',  'PGD ì‹œìˆ  ì—¬ë¶€', 'PGS ì‹œìˆ  ì—¬ë¶€', 'ê¸°ì¦ìš©', 'í˜„ì¬ ì‹œìˆ ìš©',
 'ë‚œì ì €ì¥ìš©', 'ë°°ì•„ ì €ì¥ìš©', 'ICSI', 'IVF', 'Unknown', 'IUI', 'BLASTOCYST', 'AH', 'ë‚œì„ ì—¬ë¶€', 'ìœ ì‚° ì—¬ë¶€',
 'PGS ê²€ì‚¬ë¥¼ ë°›ê³ ë„ ì•ˆí•œ ì‚¬ëŒ']

for col in cat_col:
    train_df[col] = train_df[col].astype(str)
    test_df[col] = test_df[col].astype(str)
```

```python
# ë²”ì£¼í˜• ë³€ìˆ˜ ì§€ì •
categorical_features = train_df.select_dtypes(include=['object']).columns.tolist()

# ë²”ì£¼í˜• ë³€ìˆ˜ë¥¼ category íƒ€ì…ìœ¼ë¡œ ë³€í™˜
for col in categorical_features:
    train_df[col] = train_df[col].astype("category")
    test_df[col] = test_df[col].astype("category")  # í…ŒìŠ¤íŠ¸ ë°ì´í„°ë„ ë³€í™˜
```

## ëª¨ë¸ë§

```python
# íƒ€ê²Ÿ ë¶„ë¦¬
y_train = train_df['ì„ì‹  ì„±ê³µ ì—¬ë¶€']
X_train = train_df.drop(columns=['ì„ì‹  ì„±ê³µ ì—¬ë¶€'])
X_test = test_df.copy()

## í´ë˜ìŠ¤ ê°€ì¤‘ì¹˜ ê³„ì‚°
class_counts = y_train.value_counts()
pos_weight = class_counts[0] / class_counts[1]

## ğŸ¯ K-Fold ì„¤ì •
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=sd)

## ğŸ¯ ëª¨ë¸ íŒŒë¼ë¯¸í„° ì„¤ì • (GPU ì‚¬ìš© ì ìš©)
lgb_params = {
    'objective': 'binary',
    'metric': 'auc',
    'boosting_type': 'gbdt',
    'learning_rate': 0.1,
    'num_leaves': 31,
    'random_state': sd,
    'device': 'gpu'  # GPU ì‚¬ìš©
}

cat_params = {
    'loss_function': 'Logloss',
    'eval_metric': 'AUC',
    'learning_rate': 0.1,
    'iterations': 1000,
    'depth': 6,
    'random_seed': sd,
    'od_wait': 100,
    'task_type': 'GPU',  # CPU -> GPU ë³€ê²½
    'verbose': 200,
    'scale_pos_weight': pos_weight
}

xgb_params = {
    'objective': 'binary:logistic',
    'eval_metric': 'auc',
    'learning_rate': 0.1,
    'max_depth': 6,
    'n_estimators': 1000,
    'random_state': sd,
    'scale_pos_weight': pos_weight,
    'tree_method': 'gpu_hist'  # GPU ì‚¬ìš©
}

## ğŸ¯ ëª¨ë¸ë³„ AUC ì €ì¥ êµ¬ì¡° ë³€ê²½
auc_scores_dict = {'lgbm': [], 'xgb': [], 'cat': []}
models_dict = {'lgbm': [], 'xgb': [], 'cat': []}
test_pred_dict = {'lgbm': [], 'xgb': [], 'cat': []}
```

```python
# ê° ëª¨ë¸ì—ì„œ ë“œë¡­í•  ì»¬ëŸ¼ ì§€ì •
# LGBMì—ì„œëŠ” dropí•˜ì§€ ì•ŠëŠ”ë‹¤
drop_features = {
    'lgbm': [],
    'xgb': ['PGS ê²€ì‚¬ë¥¼ ë°›ê³ ë„ ì•ˆí•œ ì‚¬ëŒ'],
    'cat': ['PGS ê²€ì‚¬ë¥¼ ë°›ê³ ë„ ì•ˆí•œ ì‚¬ëŒ']
}

# ëª¨ë¸ë³„ ë°ì´í„° ì¤€ë¹„
def prepare_model_data(model_name, X_train, X_test):
    X_train_model = X_train.drop(columns=drop_features[model_name])
    X_test_model = X_test.drop(columns=drop_features[model_name])
    categorical_features_model = [col for col in categorical_features if col not in drop_features[model_name]]
    return X_train_model, X_test_model, categorical_features_model

# LGBM ë°ì´í„° ì¤€ë¹„
X_train_lgbm, X_test_lgbm, categorical_features_lgbm = prepare_model_data('lgbm', X_train, X_test)
# XGBoost ë°ì´í„° ì¤€ë¹„
X_train_xgb, X_test_xgb, categorical_features_xgb = prepare_model_data('xgb', X_train, X_test)
# CatBoost ë°ì´í„° ì¤€ë¹„
X_train_cat, X_test_cat, categorical_features_cat = prepare_model_data('cat', X_train, X_test)
```

```python
### âœ… **5-Fold LGBM í•™ìŠµ**
for fold, (train_idx, valid_idx) in enumerate(skf.split(X_train_lgbm, y_train)):
    print(f"[LGBM] Fold {fold}: GPU device used? -> {lgb_params.get('device')}")
    X_tr, X_val = X_train_lgbm.iloc[train_idx], X_train_lgbm.iloc[valid_idx]
    y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[valid_idx]

    # LGBM ë°ì´í„°ì…‹ ìƒì„±
    lgb_train = lgb.Dataset(X_tr, label=y_tr, weight=y_tr.map({0: 1, 1: pos_weight}), categorical_feature=categorical_features_lgbm)
    lgb_valid = lgb.Dataset(X_val, label=y_val, weight=y_val.map({0: 1, 1: pos_weight}), categorical_feature=categorical_features_lgbm)

    # LGBM í•™ìŠµ
    model = lgb.train(lgb_params, lgb_train, valid_sets=[lgb_valid], callbacks=[early_stopping(100), log_evaluation(100)])
    models_dict['lgbm'].append(model)

    # ì˜ˆì¸¡ ìˆ˜í–‰
    y_pred_proba = model.predict(X_val)
    auc_score = roc_auc_score(y_val, y_pred_proba)
    auc_scores_dict['lgbm'].append((auc_score, fold))

    # í…ŒìŠ¤íŠ¸ ë°ì´í„° ì˜ˆì¸¡ ì €ì¥
    test_pred_dict['lgbm'].append(model.predict(X_test_lgbm))

print(f"LGBM ì™„ë£Œ! í‰ê·  AUC: {np.mean([x[0] for x in auc_scores_dict['lgbm']]):.5f}")
```

```python
### âœ… **5-Fold XGBoost í•™ìŠµ (DMatrixì—ì„œ ë²”ì£¼í˜• ë³€ìˆ˜ ì§ì ‘ ì‚¬ìš©)**
for fold, (train_idx, valid_idx) in enumerate(skf.split(X_train_xgb, y_train)):
    print(f"[XGBoost] Fold {fold}: GPU tree_method used? -> {xgb_params.get('tree_method')}")
    X_tr, X_val = X_train_xgb.iloc[train_idx], X_train_xgb.iloc[valid_idx]
    y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[valid_idx]

    # âœ… DMatrix ìƒì„± (enable_categorical=True)
    dtrain = xgb.DMatrix(X_tr, label=y_tr, enable_categorical=True)
    dvalid = xgb.DMatrix(X_val, label=y_val, enable_categorical=True)

    # âœ… XGBoost ëª¨ë¸ í•™ìŠµ
    model = xgb.train(
        xgb_params,
        dtrain,
        num_boost_round=1000,
        evals=[(dvalid, 'validation')],
        early_stopping_rounds=100,
        verbose_eval=100
    )

    models_dict['xgb'].append(model)

    # ì˜ˆì¸¡ ìˆ˜í–‰
    y_pred_proba = model.predict(xgb.DMatrix(X_val, enable_categorical=True))
    auc_score = roc_auc_score(y_val, y_pred_proba)
    auc_scores_dict['xgb'].append((auc_score, fold))

    # í…ŒìŠ¤íŠ¸ ë°ì´í„° ì˜ˆì¸¡ ì €ì¥
    test_pred_dict['xgb'].append(model.predict(xgb.DMatrix(X_test_xgb, enable_categorical=True)))

print(f"XGBoost ì™„ë£Œ! í‰ê·  AUC: {np.mean([x[0] for x in auc_scores_dict['xgb']]):.5f}")
```

```python
### âœ… **5-Fold CatBoost í•™ìŠµ**
for fold, (train_idx, valid_idx) in enumerate(skf.split(X_train_cat, y_train)):
    print(f"[CatBoost] Fold {fold}: GPU task_type used? -> {cat_params.get('task_type')}")
    X_tr, X_val = X_train_cat.iloc[train_idx], X_train_cat.iloc[valid_idx]
    y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[valid_idx]

    train_pool = Pool(X_tr, label=y_tr, cat_features=categorical_features_cat)
    valid_pool = Pool(X_val, label=y_val, cat_features=categorical_features_cat)

    # CatBoost í•™ìŠµ
    model = CatBoostClassifier(**cat_params)
    model.fit(train_pool, eval_set=valid_pool, early_stopping_rounds=100, verbose=100)

    models_dict['cat'].append(model)

    # ì˜ˆì¸¡ ìˆ˜í–‰
    y_pred_proba = model.predict_proba(X_val)[:, 1]
    auc_score = roc_auc_score(y_val, y_pred_proba)
    auc_scores_dict['cat'].append((auc_score, fold))

    # í…ŒìŠ¤íŠ¸ ë°ì´í„° ì˜ˆì¸¡ ì €ì¥
    test_pred_dict['cat'].append(model.predict_proba(X_test_cat)[:, 1])

print(f"CatBoost ì™„ë£Œ! í‰ê·  AUC: {np.mean([x[0] for x in auc_scores_dict['cat']]):.5f}")
```

## ìµœì¢… í•™ìŠµ ì œì¶œ

```python
## ğŸ¯ **ê°€ì¤‘ ë­í¬ ë³€í™˜ì„ ì´ìš©í•œ ì•™ìƒë¸”**
def weighted_rank_ensemble(predictions, weights):
    """ ê°€ì¤‘ ë­í¬ ë³€í™˜ì„ ì ìš©í•œ ì•™ìƒë¸” """
    predictions = np.array(predictions)
    ranked_preds = np.array([rankdata(pred) for pred in predictions])
    weighted_avg_rank = np.average(ranked_preds, axis=0, weights=weights)
    return weighted_avg_rank / np.max(weighted_avg_rank)  # ì •ê·œí™”

## ğŸ¯ ê°€ì¤‘ì¹˜ ì„¤ì •
manual_weights = {'lgbm': 1.3, 'xgb': 3, 'cat': 2.9}

## ğŸ¯ n=3ì¼ ë•Œ ê°€ì¤‘ ë­í¬ ì•™ìƒë¸” ìˆ˜í–‰
n = 3
print(f"\nğŸ”¹ í˜„ì¬ n ê°’: {n}")

selected_test_preds = []
selected_weights = []

for model_type in ['lgbm', 'xgb', 'cat']:
    # ëª¨ë¸ë³„ë¡œ AUC ë†’ì€ nê°œ ì„ íƒ
    top_n_folds = sorted(auc_scores_dict[model_type], reverse=True, key=lambda x: x[0])[:n]
    top_n_folds_idx = [fold_idx for _, fold_idx in top_n_folds]

    print(f"{model_type} ì„ íƒëœ fold: {top_n_folds_idx}")

    # ì„ íƒëœ foldì˜ ì˜ˆì¸¡ê°’ ì €ì¥
    selected_test_preds += [test_pred_dict[model_type][idx] for idx in top_n_folds_idx]

    # ì„ íƒëœ fold ê°œìˆ˜ë§Œí¼ ê°€ì¤‘ì¹˜ ì¶”ê°€
    selected_weights += [manual_weights[model_type]] * len(top_n_folds_idx)

# ê°€ì¤‘ ë­í¬ ì•™ìƒë¸” ìˆ˜í–‰
y_test_proba = weighted_rank_ensemble(selected_test_preds, selected_weights)

# ì œì¶œ íŒŒì¼ ìƒì„±
file_name = f'ensemble-weighted-rank-top{n}-per-model-fold.csv'
submission_df = pd.read_csv("./data/sample_submission.csv")
submission_df['ID'] = test_ids
submission_df.iloc[:, 1] = y_test_proba
submission_df.to_csv(file_name, index=False)

# íŒŒì¼ ë‹¤ìš´ë¡œë“œ
files.download(file_name)
```
